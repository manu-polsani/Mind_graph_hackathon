```python
import pandas as pd
import numpy as np
```

```python
meta_data=pd.read_csv("Metadata.csv")
Clubs_data=pd.read_csv("Clubs_data.csv")
Organisers_fests=pd.read_csv("Organisers_In_Fests.csv")
participants_fests=pd.read_csv("Participants_In_Fests.csv")
```

```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
Clubs_data["Best Match"]= Clubs_data["Name"].map(lambda x: process.extractOne(x,meta_data["Name"])[0])
```

```python
data=Clubs_data
data=data.drop('Name',axis=1)
data.rename(columns = {"Best Match":"Name"},inplace=True)
```

```python
data = pd.merge(data,meta_data,how="outer",on = "Name")
data.drop_duplicates(subset=['Club_Name','Event','Role','Name'], keep='first', inplace=False)
print(data)
```

```python
participants_fests["Name1"]= participants_fests["Name"].map(lambda x: process.extractOne(x,meta_data["Name"])[0])
```

```python
p=participants_fests.drop('Name',axis=1)
p.rename(columns = {"Name1":"Name"},inplace=True)
```

```python
data2 = pd.merge(data1,p,how='outer',on = "Name")
data2.rename(columns = {"Role_x":"club_role","Role_y":"Fest_role","Fest_Name_x":"oragnisers_festname","Fest_Name_y":"participants_festname","Event_x":"club_event","Event_y":"fest_event"},inplace =True)
data2.head(5)
```

```python
a=data2['Name'].unique()
a.size
```

```python
one_hot_encoded_data = pd.get_dummies(data2, columns = ['club_event','fest_event','club_role','Fest_role','oragnisers_festname','participants_festname'])
print(one_hot_encoded_data)
```

```python
one_hot_encoded_data=one_hot_encoded_data.drop_duplicates()
```

```python
l = one_hot_encoded_data["Name"].unique()
import copy 
ohc = copy.deepcopy(one_hot_encoded_data)
# one_hot_encoded_data.fillna(0,inplace=True)
# one_hot_encoded_data.replace(np.nan,0,inplace=True)
da =pd.DataFrame(np.arange(0,59).reshape(1,59),columns = list(one_hot_encoded_data.columns[1:60]))
#da

for k in range(l.shape[0]):
    for j in range(3,61):
        if(j==3):
            da.loc[k,one_hot_encoded_data.columns[j-2]] = l[k]
        if(j==4):
            da.loc[k,one_hot_encoded_data.columns[j-2]] = one_hot_encoded_data[one_hot_encoded_data["Name"] == l[k]]["ID"].iloc[0]    
        if((j>=4) and (j<59)):
            da.loc[k,one_hot_encoded_data.columns[j-1]] = one_hot_encoded_data[one_hot_encoded_data["Name"] == l[k]].iloc[:,j].sum()
        else:
           da.loc[k,one_hot_encoded_data.columns[j-1]] = one_hot_encoded_data[one_hot_encoded_data["Name"] == l[k]].iloc[:,j-1].sum()
```
